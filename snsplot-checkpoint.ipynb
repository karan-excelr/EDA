```python
import pandas as pd
import seaborn as sns
import  matplotlib.pyplot as plt
```

```python
data={"age":[22,35,45,29,36,50,24,32,40,23,48,54],
     "region":['north','south','East','West','north','south','north','West','East','south','north','East']}
```

```python
df=pd.DataFrame(data)
for region in df['region'].unique():
    subset=df[df['region']==region]
    print(f"Statistics for {region} region:")
    print(subset['age'].describe())
    # ploting the  distributuon
    sns.histplot(subset["age"],kde=True)
    plt.title(f'Age Distribution in {region} region')
    plt.show()
```
