```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data={"age":[22,35,58,45,29,36,50,24,32,40,23,48,54],"income":[32000,45000,58000,54000,36000,46000,62000,34000,39000,52000,
                                                                     33000,57000,59000]}
df=pd.DataFrame(data)
sns.scatterplot(x='age',y='income',data=df)
plt.title('age vs. income')
plt.show()
correlation=df['age'].corr(df['income'])
print(f"Correlation between age and income:{correlation}")
```

```python
df['region']=['north','south','East','West','north','East','West','south','North','West','East',
             'south','West']
sns.boxplot(x='region',y='income',data=df)
plt.title('income distribution by region')
plt.show()
sns.barplot(x='region',y='income',data=df,estimator=lambda x: sum(x)/len(x))
plt.title('Average income by Region')
plt.show()
```

```python
df['gender'] = ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female', 'Male', 'Female', 'Female', 'Male', 'Male']
#contigence Table
contingency_table=pd.crosstab(df['region'],df['gender'])
print("contingence Table")
#chi-square Test
from scipy.stats import chi2_contingency # typy: ignore
chi2,p,dof,expected= chi2_contingency(contingency_table)
print(f"chi-square Test p-value:{p}")
```
